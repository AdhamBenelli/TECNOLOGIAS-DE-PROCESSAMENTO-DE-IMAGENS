```python
import cv2
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
image_rostos = cv2.imread("rostos.jpg")
image_plt = mpimg.imread("rostos.jpg")
plt.imshow(image_plt)
plt.show()
```

```python
cv2.circle(image_rostos,(290,120), 60, (100, 50, 200), 3)
cv2.circle(image_rostos,(410,110), 60, (100, 50, 200), 3)
cv2.putText(image_rostos, "mulher", (290,200), cv2.FONT_HERSHEY_SIMPLEX, 1, (100, 50, 200), 2)
cv2.putText(image_rostos, "mulher", (410,50), cv2.FONT_HERSHEY_SIMPLEX, 1, (100, 50, 200), 2)

cv2.imwrite("rostos_processed.png", image_rostos)
image_plt = mpimg.imread("rostos_processed.png")
plt.imshow(image_plt)
plt.show()
```

```python
image_carros = cv2.imread("carros.jpg")
image_plt = mpimg.imread("carros.jpg")
plt.imshow(image_plt)
plt.show()
```

```python
cv2.rectangle(image_carros,(150,200), (480,380), (100, 50, 200), 3)
cv2.rectangle(image_carros,(0,190), (180,350), (100, 50, 200), 3)
cv2.rectangle(image_carros,(410,190), (650,350), (100, 50, 200), 3)

cv2.putText(image_carros, "volkswagem", (150,180), cv2.FONT_HERSHEY_SIMPLEX, 1, (100, 50, 200), 2)
cv2.putText(image_carros, "chevrolet", (410,170), cv2.FONT_HERSHEY_SIMPLEX, 1, (100, 50, 200), 2)
cv2.putText(image_carros, "ford", (0,170), cv2.FONT_HERSHEY_SIMPLEX, 1, (100, 50, 200), 2)

cv2.imwrite("carros_processed.png", image_carros)
image_plt = mpimg.imread("carros_processed.png")
plt.imshow(image_plt)
plt.show()
```

```python
image_placas = cv2.imread("placa-transito.jpg")
image_small = cv2.resize(image_placas, (0,0), fx=0.3, fy=0.3) 
cv2.imwrite("placa-transito-small.jpg", image_small)

image_placas = image_small.copy()

image_plt = mpimg.imread("placa-transito-small.jpg")
plt.imshow(image_plt)
plt.show()
```

```python

cv2.circle(image_placas,(725,110), 90, (100, 50, 200), 3)
cv2.circle(image_placas,(220,270), 30, (100, 50, 200), 3)

cv2.putText(image_placas, "proibido estacionar", (725,230), cv2.FONT_HERSHEY_SIMPLEX, 1, (100, 50, 200), 2)
cv2.putText(image_placas, "proibido estacionar", (220,220), cv2.FONT_HERSHEY_SIMPLEX, 1, (100, 50, 200), 2)

cv2.imwrite("placas_processed.png", image_placas)
image_plt = mpimg.imread("placas_processed.png")
plt.imshow(image_plt)
plt.show()
```
